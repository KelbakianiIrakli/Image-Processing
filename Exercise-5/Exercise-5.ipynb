```python
import cv2
import numpy as np
import matplotlib.pyplot as plt


image_directory = r'C:\Users\datog\Pictures\Screenshots\\'
image_filenames = ['pic_1.png', 'pic_2.png', 'pic_3.png', 'pic_4.png', 'pic_5.png', 'pic_6.png', 'pic_7.png', 'pic_8.png']

for filename in image_filenames:
    image = cv2.imread(image_directory + filename, cv2.IMREAD_GRAYSCALE)
    #Fourier transform
    f_transform = np.fft.fft2(image)
    f_transform_shifted = np.fft.fftshift(f_transform)
    magnitude_spectrum = np.log(np.abs(f_transform_shifted) + 1)
    plt.figure(figsize=(10, 5))
    plt.subplot(121), plt.imshow(image, cmap='gray')
    plt.title('Original Image'), plt.xticks([]), plt.yticks([])
    plt.subplot(122), plt.imshow(magnitude_spectrum, cmap='gray')
    plt.title('Fourier Transform'), plt.xticks([]), plt.yticks([])
    plt.show()
```
